```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# import chart_studio.plotly as py
# import cufflinks as cf
import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv('/content/DataPrepFinal.csv')
df.head()
```

```python

df.info()
```

```python
df.describe()
```

```python
df.drop(['Unnamed: 0','name','category','goal','pledged','usd pledged'], axis=1, inplace=True)
df.info()
```

```python
df['usd_goal_real'] = df['usd_goal_real'].str.replace("USD ","")
df['launched'] = df['launched'].str.replace(' \d\d:\d\d', '', regex=True)
```

```python
df.astype({'usd_goal_real':'float64'}).dtypes
```

```python
df.head()
```

```python
df['launched'] = pd.to_datetime(df['launched'], format='%d/%m/%y')
df['deadline'] = pd.to_datetime(df['deadline'], format='%d/%m/%y')
df['usd_goal_real'] = df['usd_goal_real'].astype('int')
df.dtypes
```

```python
df['data_range_days'] = (df['deadline'] - df['launched']).dt.days
```

```python
df.head()
```

```python
df2 = pd.read_csv('/content/campaign.csv')
```

```python
df2.head()
```

```python
df2['Text Description'].unique()
df2.drop(['Text Description'], axis=1, inplace=True)
```

```python
data = df.merge(df2, how='left', on=['ID'])
data.head()
```

```python
df3 = pd.read_csv('/content/invested.csv')
df3.head()
```

```python
df3.sort_values(by=['ID'])
```

```python
df3_dummied = pd.get_dummies(df3, columns=['backedLocation'])
df3_dummied.head()
```

```python
df3_att = df3_dummied.groupby(['ID']).agg({'age':'mean', 'backedLocation_BR':'sum', 'backedLocation_GBK' : 'sum', 'backedLocation_US':'sum'})
```

```python
data_full = data.merge(df3_att, how='right', on=['ID'])
data_full.head()
```
