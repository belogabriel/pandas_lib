```python
# libs
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# data

salarios = pd.read_csv('/content/salarios_br(2).csv', sep=',')
salarios.head(3)
```

```python
# Gráfico 1:

plt.bar
(salarios["('P5', 'living_state')"].value_counts().plot(kind='bar', title='Regiões com mais pessoas de  Dados'))
plt.show()
```

```python
# Gráfico 2:

plt.bar
(salarios["('P2', 'gender')"].value_counts().plot(kind='bar',color='lightgreen', title='Generos com mais pessoas de Dados'))
plt.show()
```

```python
# Gráfico 3:

plt.bar
(salarios["('P22', 'most_used_proggraming_languages')"].value_counts().plot(kind='bar',color='red', title='Linguagem de Prog. mais utilizadas'))
plt.show()
```

```python
# Gráfico 4:

plt.bar
(salarios["('P10', 'job_situation')"].value_counts().plot(kind='bar',color='pink', title='Empregados em carreiras de Dados'))
plt.show()
```

```python
# Gráfico 5:

plt.pie
(salarios["('D1', 'living_macroregion')"].value_counts().plot(kind='pie',labeldistance=1.3, shadow=True, title='Região com mais Profissionais em Dados'))
plt.ylabel('')
plt.show()

```
